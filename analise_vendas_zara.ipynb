```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#Importing the data

df_zara = pd.read_csv(r'C:\Users\Thomaz David\Desktop\zara\archive 2\zara.csv', sep=';')
```

```python
#visualizing the data

df_zara.head()

df_zara.columns
```

```python
#filtrando os dados 

colunas_selecionadas = ['terms', 'Sales Volume', 'price', 'section']

df_zara_filtrado = df_zara[colunas_selecionadas]

df_zara_filtrado
```

```python
#Grafico de barras de vendas por categoria de produto

quantidade_vendas_categoria = df_zara_filtrado['terms'].value_counts()

quantidade_vendas_categoria.plot(kind='bar')
plt.title('Quantidade de vendas por categoria de produto')
plt.xlabel('Categoria')
plt.ylabel('Quantidade de vendas')
plt.show()
```

```python
#Ticket Médio 

ticket_medio = df_zara_filtrado['price'].mean()

print(f'Ticket médio total: {ticket_medio:.2f}')

```

```python
#Quantidade de produtos em promoção e não promoção

produtos_promoção = df_zara['Promotion'].value_counts()

produtos_promoção.plot(kind='bar')
plt.title('Quantidade de produtos em promoção e não promoção')
plt.xlabel('Promoção')
plt.ylabel('Quantidade de produtos')
plt.show()
```

```python
#Distribuição de Preço:
#Analise da distribuição de preços dos produtos para ver quais faixas de preço são mais comuns.

df_zara_filtrado["price"].plot(kind="hist", bins=20)  # 'bins' define o número de intervalos


plt.title("Distribuição de Preço")
plt.xlabel("Faixa de Preço")
plt.ylabel("Quantidade de Produtos")
plt.show()
```

```python
#Análise de Promoção:
#Comparando as vendas entre produtos em promoção e não promocionais.

vendas_promocao = df_zara.groupby('Promotion')['Sales Volume'].sum()
vendas_promocao.plot(kind='bar', title='Vendas - Promoção vs. Não-Promoção')
plt.xlabel('Promoção')
plt.ylabel('Vendas')
plt.legend()
plt.show()
```

```python
#Variação sazonal de vendas

vendas_sazonal = df_zara.groupby('Seasonal')['Sales Volume'].sum()

vendas_sazonal.plot(kind='bar', title='Desempenho Sazonal de Vendas')
plt.xlabel('Sazonalidade')
plt.ylabel('Vendas')
plt.legend()
plt.show()

print ("Variação sazonal quase nula")

```
