```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#Importing the data

df_zara = pd.read_csv(r'C:\Users\Thomaz David\Desktop\zara\archive 2\zara.csv', sep=';')
```

```python
#visualizing the data

df_zara.head()

df_zara.columns
```

```python
#filtrando os dados 

colunas_selecionadas = ['terms', 'Sales Volume', 'price', 'section']

df_zara_filtrado = df_zara[colunas_selecionadas]

df_zara_filtrado
```

```python
#Grafico de barras de vendas por categoria de produto

quantidade_vendas_categoria = df_zara_filtrado['terms'].value_counts()

quantidade_vendas_categoria.plot(kind='bar')
plt.title('Quantidade de vendas por categoria de produto')
plt.xlabel('Categoria')
plt.ylabel('Quantidade de vendas')
plt.show()
```

```python
#Ticket MÃ©dio 

ticket_medio = df_zara_filtrado['price'].mean()

print(f'Ticket mÃ©dio total: {ticket_medio:.2f}')

```

```python
#Quantidade de produtos em promoÃ§Ã£o e nÃ£o promoÃ§Ã£o

produtos_promoÃ§Ã£o = df_zara['Promotion'].value_counts()

produtos_promoÃ§Ã£o.plot(kind='bar')
plt.title('Quantidade de produtos em promoÃ§Ã£o e nÃ£o promoÃ§Ã£o')
plt.xlabel('PromoÃ§Ã£o')
plt.ylabel('Quantidade de produtos')
plt.show()
```

```python
#DistribuiÃ§Ã£o de PreÃ§o:
#Analise da distribuiÃ§Ã£o de preÃ§os dos produtos para ver quais faixas de preÃ§o sÃ£o mais comuns.

df_zara_filtrado["price"].plot(kind="hist", bins=20)  # 'bins' define o nÃºmero de intervalos


plt.title("DistribuiÃ§Ã£o de PreÃ§o")
plt.xlabel("Faixa de PreÃ§o")
plt.ylabel("Quantidade de Produtos")
plt.show()
```

```python
#AnÃ¡lise de PromoÃ§Ã£o:
#Comparando as vendas entre produtos em promoÃ§Ã£o e nÃ£o promocionais.

vendas_promocao = df_zara.groupby('Promotion')['Sales Volume'].sum()
vendas_promocao.plot(kind='bar', title='Vendas - PromoÃ§Ã£o vs. NÃ£o-PromoÃ§Ã£o')
plt.xlabel('PromoÃ§Ã£o')
plt.ylabel('Vendas')
plt.legend()
plt.show()
```

```python
#VariaÃ§Ã£o sazonal de vendas

vendas_sazonal = df_zara.groupby('Seasonal')['Sales Volume'].sum()

vendas_sazonal.plot(kind='bar', title='Desempenho Sazonal de Vendas')
plt.xlabel('Sazonalidade')
plt.ylabel('Vendas')
plt.legend()
plt.show()

print ("VariaÃ§Ã£o sazonal quase nula")

```
